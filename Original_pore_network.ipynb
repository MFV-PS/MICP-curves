```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import porespy as ps
import inspect
from PIL import Image
import imageio.v2 as imageio
from scipy import misc
import matplotlib.patches as patches
```

```python
def pore_size (imagen, voxel, nombre_archivo):
  im = Image.open(imagen)
  thk = ps.filters.local_thickness(im,mode='dt')
  fig, ax = plt.subplots()
  ax.imshow(thk, interpolation='none', origin='lower', cmap='jet')
  ax.axis(False);

  im=thk
  data = ps.metrics.pore_size_distribution(im=im, bins=100, voxel_size=voxel, log=False)
  fig, ax = plt.subplots(1, 3, figsize=[10, 4])
  ax[0].plot(data.bin_centers,data.pdf)
  ax[1].plot(data.bin_centers,data.cdf)
  ax[2].bar(data.bin_centers, data.cdf, data.bin_widths, edgecolor='k')
  ax[0].set_title("Probability Density Function")
  ax[1].set_title("Cumulative Density Function")
  ax[2].set_title('Bar Plot')

  # Supongamos que 'data' es el objeto devuelto por la función pore_size_distribution
  bin_centers = data.bin_centers
  pdf = data.pdf
  cdf= data.cdf

  # Crear un DataFrame con bin_centers y pdf
  df = pd.DataFrame({'bin_centers': bin_centers, 'pdf': pdf, 'cdf':cdf})
  df_filtrado = df[df['pdf'] != 0]
  # Guardar el DataFrame filtrado como un archivo CSV
  df_filtrado.to_excel(nombre_archivo, index=False)


```
